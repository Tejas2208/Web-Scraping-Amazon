# Univariate Analysis

## Univariate Analysis – Brand Distribution
Analyzing frequency of sneaker brands to identify which brand dominates the dataset.

```python
plt.figure(figsize=(15, 6))
sns.countplot(data = df,x = 'Title')
plt.title('Frequency of Sneaker Brands')
plt.xlabel('Brands')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Observations:
1. PUMA has the highest count in the dataset more than 60.
2. Following PUMA,brands like Nike, Campus and U.S. Polo Assn. are showing higher frequency than rest of the brands.
3. Most of the brands like Levi's, FILA, Mochi, JQR appear only for handful of times.

## Price Distribution
Analyzing distribution of original prices to understand:
- Price range
- Skewness
- Common price brackets

```python
# Price Distribution

sns.histplot(df['Original Price'], bins= 30, kde= True)
plt.title('Price Distribution')
plt.show()
```

### Observations:
1. Most products are priced at range from Rs.3000 to Rs.5000. The peak density is clearly around Rs.4000
2. The distribution of the Sneaker products are right-skewed as tail is tilted towards higher prices, very few brands products exist there.
3. Amazon sneaker market is heavily concentrated in mid-range pricing, with very limited premium presence.

```python
# Rating Distribution
plt.figure()
sns.histplot(df['Rating'], bins = 10)
plt.title('Rating Distribution')
plt.show()
```

### Observations
1. Ratings are heavily concentrated between 3.8 and 4.5, that means almost everything is rated 'good'
2. the distribution left skewed because very few products fall below 3.We can interpret from this that poor ratings don't get visibility and reviews.
3. Top-rated products are very rare, that means they must be premium brand sneakers.

```python
# Boxplot - Discount %
plt.figure()
sns.boxplot(x=df['Discount %'])
plt.title("Discount % Spread")
plt.show()
```

### Observations
1. Median discount is around 48-50%. That may seen as a huge discount, but in reality, in platforms like Amazon, Original price of the product is set higher than it's actual value to attract the customer by showing a big discount like 50%.
2. Most discounts lie between 35% to nearly 60%. Anything outside this range is either under-discounted or aggresively pushed.
3. There are no strong outliers visible. Discounts are consistent, not dominated by extreme outliers.This tells that discounting standard pricing strategy across Sneakers rather than any promotional tactic.

# Bivariate Analysis

## Bivariate Analysis – Price vs Rating
Examining relationship between product price and customer ratings to identify if higher-priced products receive better ratings.

```python
# Price vs Rating
sns.regplot(x='Original Price', y='Rating', data=df)
plt.title("Price vs Rating")
plt.show()
```

### Observations
1. Points are spread randomly across the graph, that means there's no strong relationship between 'Price' and 'Rating'.
2. Maximum Ratings are stucked bewtween 3.8 to 4.3 even the Sneaker is cheap or expensive, this tells that price has negligible impact on Ratings.
3. Expensive Sneaker Brands are not consistently better, you can clearly see that Expensive ones has average Ratings where Mid price Sneakers have High Ratings.
4. The scatter plot shows no strong relationship bewtween Price and Ratings, indicating that higher priced sneakers do not necessarily receive better customer ratings.

```python
# Popularity vs Number of reviews
sns.barplot(x=df['Popularity'], y=df['Review Count'])
plt.xlabel('Popularity')
plt.ylabel('Review Count')
plt.title('Popularity based on number of reviews')
plt.show()
```

### Observations
1. Highly Popular Brands dominate in review count as it has the largest average review count, around 700.
2. The Moderate category is around 270, which is much lower than the high group. So popularity does not rise gradually—it jumps sharply.
3. The Low group is near 30 reviews, showing these products get very little customer attention and weak market visibility.
4. This indicates that customer attention is heavily concentrated on a small set of products, while most products remain largely unnoticed

```python
# Discount % vs Review Count
plt.figure()
sns.scatterplot(x='Discount %', y='Review Count', data=df)
plt.title("Discount vs Reviews")
plt.show()
```

### Observations:
1. Points are scattered randomly, this shows no clear relationship bewtween Discount and Reviews.
2. Most products have very less number of reviews regardless of the discounts, this means majority of the products get minimal engagement.
3. Offering higher discounts does not guarantee higher popularity by looking this plot.

```python
sns.violinplot(data=df,x=df['Rating'],y = df['Popularity'])
```

# Top 5 most reviewed products

```python
top_products = df.sort_values(by='Review Count', ascending=False).head(10)

sns.barplot(x='Review Count', y='Title', data=top_products)
plt.title("Top Reviewed Sneakers")
plt.show()
```

### Observations
1. The difference between #1 and #5 is very minimal, this tells very strong competetion among Top brands. No single product is overwhelmingly dominates.
2. Multiple brands are appearing in the top lost. That means leadership is shared, not monopolized
3. Top products likely benefit from visibility or Brand's trust, not just on pricing or discounts.

```python
top_brands = df['Title'].value_counts().head(10)

sns.barplot(x=top_brands.values, y=top_brands.index)
plt.title("Top Brands")
plt.show()
```

### Observations
1. PUMA clearly leads the product count as it has the highest number of listing among all brands.
2. Campus is the second strongest playes just behind PUMA but still significantly ahead of others.
3. Mid-tier brands like U.S. Polo Assn., SPARX,Lotto have moderate presence, Competitive but not dominant.
4. The rest are clustered together, meaning there is no strong differentiation in listing volume.
5. A few brands dominate listings, while many others have limited presence. Market is skewed, not evenly distributed.

```python
brand_rating = df.groupby('Title')['Rating'].mean().sort_values(ascending=False).head(10)

sns.barplot(x=brand_rating.values, y=brand_rating.index)
plt.title("Best Rated Brands")
plt.show()
```

### Observations:
1. Levi's has the highest average rating, it looks like a top performer but this could be based on very few products of it.
2. Most brands like Lotto, Boldfit, Bata, Campus, etc. are tightly clustered. There are very small differences, not significant.
3. Popular brands like PUMA or Nike are not leading here. This tells us that high popularity does not guarentee higher ratings.
4. Lesser-known brands like Centrino presence in the top list is likely due to fewer products or niche offerings.

# Multivariate Analysis

## Correlation Heatmap Analysis

This heatmap shows the correlation between numerical features such as:
- Discounted Price
- Original Price
- Rating
- Review Count
- Discount %

```python
plt.figure()
corr = df[['Discounted Price', 'Original Price', 'Rating', 'Review Count', 'Discount %']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

### Obsevations
1. There is strong positive correlation between Original Price & Discounted Price is seen. That means expensive products remain expensive even after discounting.
2. We can see moderate negative correlation between Discount % & Discounted Price. It is expected that higher discounts generally lead to lower final prices.
3. Very weak correlation between Price and Rating. Price has almost no impact on customer ratings.
4. Very weak correlation between Rating and Review Count, that means highly rated are not necessarily popular.
5. Slight positive correlation between Discount % and Review Count. Discounts may slightly improve engagement.
6. Negative correlation between Discount % and Rating. Heavily discounted products tend to have slightly lower ratings.
